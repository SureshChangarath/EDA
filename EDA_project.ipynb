# What is Obesity?

According to the World Health Organization (WHO), obesity is a chronic disease that's defined by excessive fat accumulation that can harm health. It's classified by body mass index (BMI), waist circumference, and disease risk. A BMI of 25 or higher is considered overweight, and 30 or higher is considered obese.

Factors that can cause obesity include:
* Eating patterns
* Lack of sleep or physical activity
* Some medicines
* Family inheritance
* Unhealthy diet
* Social and economic issues
* Age

The data used in this analysis contains information collected from Peru, Mexico and Colombia. It was retrieved from "UC Irving Machine Learning Repository". 

The dataset information described below: 

* Gender — Female or Male

* Age — Age of respondent. This ranges between 14 and 61

* Height — Height of respondent in meters

* Weight — weight of respondent in kilograms

* BMI —  weight/(height)2 

* family_history_with_overweight — yes or no

* High_calorie_Food — if respondent eats high calorie foods frequently(yes or no)

* Eat_Vegetables — Does respondent eat vegetables with meals (never, sometimes or always.)

* Meals_per_day — how many meals per day(1 or 2, 3, >3)

* Eat_beteen_meals — Eat between meals?(no, sometimes frequently or always)

* Smoke — does respondent smoke(yes or no)

* Water_Consumption — How much water per day does respondent drink in liters?(<1, between 1 and 2, >2)

* Daily_Calorie_Count — do you count daily calories(yes or no)

* Excercise_Count — number of days of physical activity(0 = none, 1 = 1 to 2 days,2= 2 to 4 days, 3 = 4 to 5 days etc)

* Tech_Use — hours of technology use(0 = 0–2 hours, 1 = 3–5 hours, 2 = more than 5 hours)

* Alchohol — alcohol consumption(never, sometimes, frequently or always)

* Transportation — means of transportation(automobile, motorbike, bike, public transportation, or walking

* Obesity_Condition — obesity classification(insufficient weight, normal weight, level I overweight, level II overweight, type I obesity, type 2 obesity, type 3 obesity)

# Which factors have direct correlation with BMI? How does increase in age affects BMI?

```python
print(df.corr()) 
  
# plotting correlation heatmap 
dataplot = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True) 
  
# displaying heatmap 
plt.show() 
```

A correlation value of 0.93 is close to 1, which suggests a very strong positive linear relationship between BMI and weight. This indicates that as weight increases, BMI also tends to increase, and vice versa.In practical terms, this correlation value implies that weight is a strong predictor or indicator of BMI. Individuals with higher weights are likely to have higher BMIs, and those with lower weights are likely to have lower BMIs.With such a high correlation, weight can be considered a reliable predictor of BMI. Therefore, if you know someone's weight, you can predict their BMI with a high degree of accuracy using a linear relationship.

Increasing age can affect BMI (Body Mass Index), but the relationship between age and BMI is not straightforward and can vary depending on various factors like Changes in Body Composition,Lifestyle Factors,Health Conditions, etc.

# What is the range of BMI for good obesity condition(normal weight)?

```python
plt.figure(figsize=(15,6)) 
sns.boxplot(x='Obesity_Condition', y='BMI', data=df)
plt.title("Box plot")
plt.show()
```

From the above boxplot we can found that BMI range above 2O and below 25 is the optimum BMI for a healthy person.Between 18.5 and 24.9 is the optimum BMI range for a healthy adult. Additionally, BMI ranges may not be applicable to certain populations, such as athletes or older adults, who may have different body compositions or health considerations.
Therefore, it's essential to consider BMI in conjunction with other factors, such as waist circumference, body composition, overall health status, and lifestyle factors, to assess an individual's health risk accurately.

Ultimately, the optimum BMI range for humans depends on various factors, including age, sex, ethnicity, and individual health goals. It's essential to consult with a healthcare professional to determine an appropriate target BMI range based on individual circumstances and health considerations.

# Does physical activities and a person’s mode of transport have any impact on obesity?

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Obesity_Condition', y='Exercise_Count', data=df)
plt.title("Bar plot")
plt.show()
```

For the data we are given, people involving in physical activity can maintain good obesity level.Physical activity plays a crucial role in preventing and managing obesity by promoting energy balance,encouraging healthy lifestyle behaviors,etc.But it is also clear along with physical activity e have to balance other factors affecting high obesity for good health. 

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Obesity_Condition', hue='Transportation', data=df)
plt.title("Count plot")
plt.show()
```

The lowest obesity levels is seen in category of people who walk when compared with other modes of transport.Other mode of transportation shows variabilities may be due to other lifestyle choices. If walking and biking are considered as physical activities, then, it should not be surprising that people in these category have lower median BMIs.

# Does consumption of high calorie foods affect obesity?

```python
plt.figure(figsize=(15,6))
sns.barplot(x='High_Calorie_Food', y='BMI', data=df)
plt.show()
```

Consuming high-calorie foods regularly can have a significant impact on BMI by promoting weight gain, altering body composition, increasing the risk of obesity, and influencing metabolic health. Maintaining a balanced diet that includes a variety of nutrient-dense foods and practicing portion control can help support a healthy BMI and overall well-being.
